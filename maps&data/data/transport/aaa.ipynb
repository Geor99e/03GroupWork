```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas 
import json
```

```python
lines22 = gpd.read_file("./tfl_lines.json")
lines22
```

```python
print(type(lines22["lines"].iloc[0]))
print(lines22["lines"].iloc[0])
```

```python
# 将 JSON 字符串转换为 Python 对象（列表 of 字典）
def parse_lines(val):
    try:
        return json.loads(val)
    except json.JSONDecodeError:
        return []  # 如果解析失败，则返回空列表

# 应用转换
lines22["lines_parsed"] = lines22["lines"].apply(parse_lines)

# 获取所有不同的线路名称
all_names = set()
for line_list in lines22["lines_parsed"]:
    for entry in line_list:
        name = entry.get("name")
        if name:
            all_names.add(name)

print(sorted(all_names))
```

```python
# 为每个 name 创建一个新列
for name in all_names:
    lines22[name] = lines22["lines_parsed"].apply(
        lambda line_list: int(any(entry.get("name") == name for entry in line_list))
    )
```

```python
stations = gpd.read_file('./tfl_stations.json')
stations
```

```python
# 转换 lines 字段为列表
def parse_station_lines(val):
    try:
        return json.loads(val)
    except:
        return []

stations["lines_parsed"] = stations["lines"].apply(parse_station_lines)
```

```python
# 收集所有 station line 名称
station_line_names = set()
for line_list in stations["lines_parsed"]:
    for entry in line_list:
        name = entry.get("name")
        if name:
            station_line_names.add(name)

print(sorted(station_line_names))
```

```python
# 创建多个布尔列，表示站点是否属于某条线路
for name in station_line_names:
    stations[name] = stations["lines_parsed"].apply(
        lambda line_list: int(any(entry.get("name") == name for entry in line_list))
    )
```

```python
# 5. 导出为 GeoJSON（会保留投影信息）
stations.to_file("london_stations.geojson", driver="GeoJSON")
```
